# Getting started with GTFS
## What is GTFS?
GTFS(General Transit Feed Specification) is data specification that allows companies to share data about public vehicle, transits, stop for a large variaty of software.

Example of usage:
![](https://c8.alamy.com/comp/RKG9K3/train-station-departure-board-strike-cancel-train-france-french-paris-gare-du-nord-RKG9K3.jpg)
